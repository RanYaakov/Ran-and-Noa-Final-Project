```python
import bs4
from bs4 import BeautifulSoup  
import pandas as pd
import scipy as sc
import numpy as np
import requests
```

```python
pages=20
names=list()
final_urls=list()
base_url="https://www.metacritic.com/"
for page in range(0,pages):
    url="https://www.metacritic.com/browse/games/score/metascore/all/ps4/filtered?page="+str(page)
    user_agent={'User-agent':'Chrome'}
    html = requests.get(url,headers=user_agent)
    soup=BeautifulSoup(html.content,"html.parser")

    tbl = soup.find_all('table')
    
    for i in range(0,4):
        res=tbl[i].find_all('h3')
        if i<3:
            for j in range(0,5):
                names.append(res[j].text)
                
        else:
            if page<pages:
                for j in range(0,85):
                    names.append(res[j].text)
            else: 
                for j in range(0,58):
                    names.append(res[j].text)

                    
    urls=soup.find_all("a",attrs={"class":"title"})
    if page<pages:
        for i in range(0,100):
            final_urls.append(base_url+urls[i]['href'])
        
    else:   
        for i in range(0,73):
            final_urls.append(base_url+urls[i]['href'])

        



```

```python
publishers=list()
date=list()
platforms=list()
metascores=list()
users_scores=list()
players=list()
genres=list()
raiting=list()
developers=list()
for url in final_urls:
    user_agent={'User-agent':'Chrome'}
    html = requests.get(url,headers=user_agent)
    soup=BeautifulSoup(html.content,"html.parser")
    try:
        x=soup.find('span',attrs={"class":"data"}).find_all('a')
        pub=list()
        if len(x) == 1:
            publishers.append(x[0].text.strip())
        else:
            pub=list()
            for i in range(len(x)):
                pub.append(x[i].text.strip())
            publishers.append(pub)
    
    except:
        publishers.append(None)
    
    try:
        date.append(soup.find('li', attrs={"class":"summary_detail release_data"}).find('span',attrs={"class":"data"}).text)
    except:
        date.append(None)
    
    try:
        plat=soup.find('li',attrs={"class":"summary_detail product_platforms"})
        if "PC" not in plat.text and "Xbox One" not in plat.text:
             platforms.append(None)
            
        else:
            plat=plat.find_all('a')
            p=list()
            for i in range(len(plat)):
                if "PC" in plat[i].text or "Xbox One"  in plat[i].text:
                    p.append(plat[i].text)
        
            platforms.append(p)
    except:
        platforms.append(None)
    try:        
        rate=soup.find_all('span',itemprop="ratingValue")
        metascores.append(rate[0].text)
    except:
        metascores.append(None)
    try:
        users=soup.find('div',attrs={"class":"userscore_wrap feature_userscore"}).find('a')
        users_scores.append(users.text.strip())
    except:
        users_scores.append(None)
    try:
        genre=soup.find('li',attrs={"class":"summary_detail product_genre"}).find_all('span',attrs={"class":"data"})
        g=list()
        for i in range(0,len(genre)):
            g.append(genre[i].text)
        genres.append(g)
    except:
        genres.append(None)
    try:
        numOf=soup.find('li',attrs={"class":"summary_detail product_players"}).find('span',attrs={"class":"data"})
        players.append(numOf.text)
    except:
        players.append(None)
    try:
        rate=soup.find('li',attrs={"class":"summary_detail product_rating"}).find('span',attrs={"class":"data"})
        raiting.append(rate.text)
    except:
        raiting.append(None)
    try:
        dev=soup.find('li',attrs={"class":"summary_detail developer"}).find_all('a')
        if len(dev) == 1:
            developers.append(dev[0].text.strip())
        else:
            d=list()
            for i in range(len(dev)):
                d.append(dev[i].text.strip())
            developers.append(d)
    
    except:
        publishers.append(None)
```

```python

df=pd.DataFrame({"name":names,"release":date,"genres":genres,
                    "metascore":metascores,"users_score":users_scores,"platforms":platforms,"publisher":publishers,
                "num_of_players":players,"rating":raiting,"developer":developers})

#df.to_csv('ps4.csv', encoding='utf-8',index=False)



```
